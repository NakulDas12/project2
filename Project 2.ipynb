```python
import pandas as pd
import numpy as py


dataset_1 = pd.read_excel('/Users/nakuldas/Downloads/dataset_1.xlsx')
dataset_2 = pd.read_excel('/Users/nakuldas/Downloads/dataset_2.xlsx')

dataset_2['atemp'] = dataset_2['atemp'].fillna(dataset_2['atemp'].mean())

combined_data = pd.merge(dataset_1, dataset_2, on='instant')

print("Combined Data Shape:", combined_data.shape)
print("Combined Data Info:")
print(combined_data.info())


combined_data.drop(columns=['Unnamed: 0'], inplace=True)
unique_season = combined_data['season'].unique()
unique_holiday = combined_data['holiday'].unique()
unique_weathersit = combined_data['weathersit'].unique()

print("Unique Seasons:", unique_season)
print("Unique Holidays:", unique_holiday)
print("Unique Weather Situations:", unique_weathersit)

print("Data Types:\n", combined_data.dtypes)

missing_values = combined_data.isnull().sum()
print("Missing Values:\n", missing_values)

combined_data['dteday'] = pd.to_datetime(combined_data['dteday'])
print("Data Types after conversion:\n", combined_data.dtypes)

combined_data['year'] = combined_data['dteday'].dt.year
combined_data['month'] = combined_data['dteday'].dt.month
combined_data['day'] = combined_data['dteday'].dt.day

print(combined_data.describe())
correlation = combined_data.corr()
print(correlation)

import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = combined_data.select_dtypes(include=['number'])

Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
outliers = numeric_cols[(numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))]
print(outliers)

combined_data.to_excel('processed_combined_data.xlsx', index=False)


print(combined_data['temp'].skew())
correlation_matrix = combined_data.corr()
print(correlation_matrix)

```
